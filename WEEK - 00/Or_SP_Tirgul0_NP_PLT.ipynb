````python
# create multiple subplot objects 
fig, (ax1, ax2) = plt.subplots(nrows=2,ncols=1,figsize=(10, 10))

# Plot the first graph in the upper subplot
ax1.plot(x, np.exp(x));
# Plot the another curve on the same subplot
ax1.plot(x, np.log(x),c = 'yellow');

# Set title, labels to the x and y axis - ALWAYS ADD TITLES!!!!
ax1.set(title="logaritem plots", xlabel="x-axis", ylabel="y-axis");

# Add legend
ax1.legend(["exp","log"]);

# Use "scatter" plots to plot a sin and cosine wave
ax2.scatter(x, np.sin(x*2*np.pi),c = "red");
ax2.scatter(x, np.cos(x*2*np.pi),c = "green");

# Set title, labels to the x and y axis - ALWAYS ADD TITLES!!!!
ax2.set(title="logaritem plots",xlabel="time[sec]", ylabel="amp");
````

### Writing a function

Whenever you are writing a function, you need to think of the following things:

* What is the purpose of the function?
* How should I name the function?
* What input does the function need?
* What output should the function generate?


### Why use a function?

There are several good reasons why functions are a key component of any non-ridiculous programmer:

* encapsulation: wrapping a piece of useful code into a function so that it can be used without knowledge of the specifics

* generalization: making a piece of code useful in varied circumstances through parameters

* manageability: Dividing a complex program up into easy-to-manage chunks

* maintainability: using meaningful names to make the program better readable and understandable

* reusability: a good function may be useful in multiple programs

````python
def element_multiply(x, y):
    """Multiply two vectors element by element. (need to be same length)"""
    loop = np.arange(len(x))
    result = loop  
    if len(x) != len(y):
      print("your vectors must be of same length")
      return 0
    else:
      for i in loop:
        result[i] = (x[i]*y[i])
      return (result,3)
````

### build a function to plot a sine wave with a given cycle and amplitude 

````python
import numpy as np
import matplotlib.pyplot as plt

def plot_sine_wave(cycles, amplitude):
    # Generate x values for one cycle of the sine wave
    x = np.linspace(0, 2*np.pi, 100)

    # Generate y values for the sine wave with the given amplitude
    y = amplitude * np.sin(cycles*x)


    fig, (ax1) = plt.subplots(nrows=1,ncols=1,figsize=(10, 10))

    # Plot the sine wave
    plt.scatter(x, y)
    plt.xlabel('radians')
    plt.ylabel('Amplitude')
    plt.title(f'Sine Wave with {cycles} Cycles and Amplitude of {amplitude}')
    plt.show()
````

````python
plot_sine_wave(5, 4)
````
