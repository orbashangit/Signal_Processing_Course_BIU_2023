# Trigonometric Functions (amplitude, phase and frequency)

$ y = Amplitude * sin(frequency*(duration * 2\pi) + phase) + verticalShift $

```python


def plot_sine_and_cosine_wave(amplitude,phase,frequency,duration,sampling_rate):
    # Generate t values for the given duration (in seconds) and given sampling_rate(per seconds) of the sine wave
    
    t = np.arange(0, duration, 1/sampling_rate)

    # Generate y values for the sine wave with the given amplitude frequency and phase
    ysin = amplitude * np.sin(frequency*t*(2*np.pi)+phase)

    # Generate y values for the cosine wave with the given amplitude frequency and phase
    ycos = amplitude * np.cos(frequency*t*(2*np.pi)+phase)

    # create multiple subplot objects 
    fig, (ax) = plt.subplots(nrows=1,ncols=1,figsize=(16, 8))

    # Plot the "scatter" graph of both Sine and Cosine waves
    # Set title, labels to the x and y axis of the sine plot
    # - ALWAYS ADD TITLES!!!!

    ax.plot(t, ysin, 'r-o')
    #ax.scatter(t, ycos, c = 'blue');
    ax.set(title=(f'Waves with Amplitude of: {amplitude}, phase shift: {phase} [radians], frequency {frequency}[Hz], duration {duration}[sec] and sampling rate of {sampling_rate}[Hz]'), xlabel="Time (sec)", ylabel="Amplitude");
    ax.legend(["Sine"])
    #ax.legend(["Sine", "Cosine"])

```

```python
# import
import numpy as np
import matplotlib.pyplot as plt
from numpy.core.arrayprint import array2string

fig, (ax) = plt.subplots(nrows=1,ncols=1,figsize=(10, 10))

#time vector
t = np.arange(0, 1, 0.001)

amplitudeVec = np.array([1, 2, 3, 4])
# create a sine with different amplitude in each iteration
for i in range(len(amplitudeVec)):
    ySin = amplitudeVec[i] * np.sin(2*np.pi*t)
    ax.plot(t, ySin, '-o')



ax.set(title="Sine waves with different amplitudes", xlabel="Time (sec)", ylabel="Amplitude");
ax.legend([np.array2string(amplitudeVec[0]), np.array2string(amplitudeVec[1]), np.array2string(amplitudeVec[2]), np.array2string(amplitudeVec[3])])

```

## combining two sine wave with diffrent frequencies

```python
duration = 3 # 3 seconds 
highSamplingRate= 1000 # approximation to original signal
frequency3 = 3
frequency7 = 7


t = np.arange(0, duration, 1/highSamplingRate)
y = np.sin(frequency3*t*(2*np.pi)) + np.sin(frequency7*t*(2*np.pi))

fig, (ax) = plt.subplots(nrows=3,ncols=1,figsize=(20, 10))
# 3 Hz
ax[0].plot(t, np.sin(frequency3*t*(2*np.pi)),'-o')
ax[0].set(title=(f'Waves with frequency {frequency3}[Hz]'), xlabel="Time (sec)", ylabel="Amplitude");
# 7 Hz
ax[1].plot(t, np.sin(frequency7*t*(2*np.pi)),'-o')
ax[1].set(title=(f'Waves with frequency {frequency7}[Hz]'), xlabel="Time (sec)", ylabel="Amplitude");
# 3 Hz + 7Hz
ax[2].plot(t, np.sin(frequency3*t*(2*np.pi)) + np.sin(frequency7*t*(2*np.pi)),'-o')
ax[2].set(title=(f'Waves with frequency {frequency3}[Hz] + frequency {frequency7}[Hz]'), xlabel="Time (sec)", ylabel="Amplitude");

fig.tight_layout() # helps with spacing subplots
```

```python
duration = 3 # 3 seconds 
frequency3 = 3#Hz
frequency7 = 7#Hz


highSamplingRate = np.arange(0, duration, 1/1000) # approximation to original signal

normalSamplingRate = np.arange(0, duration, 1/100)

minSamplingRate = np.arange(0, duration, 1/15) # more then 7*2

lowSamplingRate = np.arange(0, duration, 1/8) # less then 7*2



yHigh = np.sin(frequency3*highSamplingRate*(2*np.pi)) + np.sin(frequency7*highSamplingRate*(2*np.pi))
yNormal = np.sin(frequency3*normalSamplingRate*(2*np.pi)) + np.sin(frequency7*normalSamplingRate*(2*np.pi))
yLow = np.sin(frequency3*lowSamplingRate*(2*np.pi)) + np.sin(frequency7*lowSamplingRate*(2*np.pi))
yMin = np.sin(frequency3*minSamplingRate*(2*np.pi)) + np.sin(frequency7*minSamplingRate*(2*np.pi))


fig, (ax) = plt.subplots(nrows=4,ncols=1,figsize=(20, 10))
# 1000 Hz
ax[0].plot(highSamplingRate, yHigh,'-o')
ax[0].set(title=(f'Waves with Sampling Rate of 1000[Hz]'), xlabel="Time (sec)", ylabel="Amplitude");
# 100 Hz
ax[1].plot(normalSamplingRate, yNormal,'-o')
ax[1].set(title=(f'Waves with Sampling Rate of 100[Hz]'), xlabel="Time (sec)", ylabel="Amplitude");
# 3 Hz + 7Hz
ax[2].plot(minSamplingRate,yMin ,'-o')
ax[2].set(title=(f'Waves with Sampling Rate of 15[Hz]'), xlabel="Time (sec)", ylabel="Amplitude");

ax[3].plot(lowSamplingRate, yLow,'-o')
ax[3].set(title=(f'Waves with Sampling Rate of 8[Hz]'), xlabel="Time (sec)", ylabel="Amplitude");

fig.tight_layout() # helps with spacing subplots
```
