```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create vectors of zeros and 4s (pulse)
zeros = np.zeros((100))
pulse = 4*np.ones((50))
#pad the pulse with zeros from both sides using concatenate
y = np.concatenate((zeros, pulse, zeros))
# create the time vector of same length
t = np.arange(0, np.size(y), 1)  
# generate randon noise (normally distributed)
noise = np.random.randn(np.size(y))
# add noise to the generated pulse signal
noisySignal = y+noise

fig, (ax) = plt.subplots(nrows = 3, ncols = 1, figsize=(16, 8))

ax[0].plot(t, y)
ax[0].set(title="Raw signal", xlabel="Times (sec)", ylabel = "Amplitude")

ax[1].plot(t, noise)
ax[1].set(title="Normally distributed noise", xlabel="Times (sec)", ylabel = "Amplitude")

ax[2].plot(t, noisySignal)
ax[2].set(title="Noisy signal", xlabel="Times (sec)", ylabel = "Amplitude")

fig.tight_layout()
```

Calculating SNR

```python
import numpy as np

# generate a noisy signal
loc = slice(100, 150)# defined as index 100:149
signal = np.concatenate((np.zeros(100), 8* np.ones(50), np.zeros(100)))
noise = np.random.randn(len(signal))
noiseSignal = signal + noise

# power of the noise and the signal
signal_power = np.sum(signal[loc] ** 2) / len(signal[loc])
noise_power = np.sum(noise[loc] ** 2) / len(noise[loc])

# compute the SNR
#power
SNR_power = signal_power / noise_power
print(SNR_power)
#RMS
SNR_rms = np.sqrt(signal_power / noise_power)
print(SNR_rms)


```

```python
# Alternative definitions:

#mean(signal)/std(noise)
SNR_alt_def = np.mean(signal[loc]) / np.std(noise[loc])
#mean(signal)/std(signal)
SNR_alt_def_2 = np.mean(noiseSignal[loc]) / np.std(noiseSignal[loc])
```
