DFT example #1 - single frequency

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate the signal (10Hz sine wave)
T = 1  # signal duration (sec)
Fs = 42  # Hz
N = T*Fs

# Create the sine wave
t = np.arange(0, T, 1/Fs)
sineFrequency = 20 #Hz
signal = 5*np.sin(2 * np.pi * sineFrequency * t)

# 1. FFT
X = np.fft.fft(signal)

# 2. Normalize DFT coefficients by N/2
normalized_X = X/(N/2)

# 3. Take the absolute values of X (the DFT)
normalized_X = np.abs(normalized_X)


# Determine the corresponding frequencies
frequencyVec = np.arange(0,(N)/T, 1/T)


# Plot the signal and DFT coefficients
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# Plot the original signal
axs[0].plot(t, signal, 'b.-')
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Amplitude')
axs[0].set_title('Original Signal (10Hz Sine Wave)')

# Plot the DFT coefficients centered around the Nyquist frequency
axs[1].stem(frequencyVec, normalized_X)
axs[1].set_xlabel('Frequency (Hz)')
axs[1].set_ylabel('Magnitude')
axs[1].set_title('DFT Coefficients (Centered at Nyquist)')

plt.tight_layout()
plt.show()
```

DFT example #2 - multiple frequencies

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate the signal (10Hz sine wave)
T = 1.0  # signal duration
Fs = 81  # Hz
N = T*Fs

t = np.arange(0, T, 1/Fs)
sineFrequency1 = 3
sineFrequency2 = 10
signal = np.sin(2 * np.pi * sineFrequency1 * t) + np.sin(2 * np.pi * sineFrequency2 * t)+5

# FFT
dft = np.fft.fft(signal)

# Normalize DFT coefficients by N/2
normalized_dft = dft/(N/2)

# Take the absolute values of X (the DFT)
normalized_dft = np.abs(normalized_dft)

# Determine the corresponding frequencies
frequencyVec = np.arange(0,(N)/T, 1/T)


# Plot the signal and DFT coefficients
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# Plot the original signal
axs[0].plot(t, signal, 'b.-')
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Amplitude')
axs[0].set_title('Original Signal (10Hz Sine Wave)')

# Plot the DFT coefficients centered around the Nyquist frequency
axs[1].stem(frequencyVec, normalized_dft)
axs[1].set_xlabel('Frequency (Hz)')
axs[1].set_ylabel('Magnitude')
axs[1].set_title('DFT Coefficients (Centered at Nyquist)')

plt.tight_layout()
plt.show()
```
